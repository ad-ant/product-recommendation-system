## Dataset size
The order_products_prior table contains over 30 milion rows, whitch can slow down exploratory analysis and experimentation.

To make development faster, a random subset of the dataset is used during the initial stages of the project.

Using a sample allows for:
- faster experimentation and debugging
- lower memory usage
- preserving the overall data distribution through random sampling

The full dataset can later be used for final model training if needed.

Sample size used in this notebook: 200 000 rows.

```python
order_products = order_products.merge(
    products[['product_id', 'product_name']],
    on='product_id'
)
```

```python
order_products.head()
```

```python
order_products.columns
```

```python
order_products_sample = order_products.sample(n=200000, random_state=42)
```

### **Processed dataset**
After merging product information and sampling the data, a processed dataset was created containing the following fields:
- order_id,
- product_id,
- add_to_cart_order,
- reordered,
- product_name.

This dataset will serve as the input for building the recommendation system.
